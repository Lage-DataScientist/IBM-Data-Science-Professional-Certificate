```python
launch_sites = pd.DataFrame({
    'Launch Site': [
        'CCAFS LC-40',
        'VAFB SLC-4E',
        'KSC LC-39A',
        'CCAFS SLC-40'
    ],
    'Lat': [28.562302, 34.632834, 28.573255, 28.561857],
    'Lon': [-80.577356, -120.610746, -80.646895, -80.577366]
})

# Cria um mapa centrado no meio dos sites
launch_map = folium.Map(location=[launch_sites['Lat'].mean(), launch_sites['Lon'].mean()], zoom_start=4)

# Adiciona os marcadores
for _, row in launch_sites.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Lon']],
        popup=row['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(launch_map)

# Mostra o mapa
launch_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
import pandas as pd
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

# Função para calcular distância
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

# Lista de sites com coordenadas
launch_sites = [
    {
        'site': 'CCAFS LC-40',
        'lat': 28.562302,
        'lon': -80.577356,
        'coast_lat': 28.56367,
        'coast_lon': -80.57163
    },
    {
        'site': 'CCAFS SLC-40',
        'lat': 28.563197,
        'lon': -80.576820,
        'coast_lat': 28.56367,
        'coast_lon': -80.57163
    },
    {
        'site': 'KSC LC-39A',
        'lat': 28.573255,
        'lon': -80.646895,
        'coast_lat': 28.57380,
        'coast_lon': -80.63960
    },
    {
        'site': 'VAFB SLC-4E',
        'lat': 34.632834,
        'lon': -120.610746,
        'coast_lat': 34.63710,
        'coast_lon': -120.60650
    }
]

# Mapa base
site_map = folium.Map(location=[28.5, -80.6], zoom_start=5)

# Adicionar marcadores e linhas
for site in launch_sites:
    distance = calculate_distance(site['lat'], site['lon'], site['coast_lat'], site['coast_lon'])
    
    # Launch site marker
    folium.Marker(
        location=[site['lat'], site['lon']],
        popup=f"{site['site']}",
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(site_map)
    
    # Coastline point marker
    folium.Marker(
        location=[site['coast_lat'], site['coast_lon']],
        popup="Closest Coastline",
        icon=folium.Icon(color='green')
    ).add_to(site_map)

    # Linha entre o launch site e a costa
    folium.PolyLine(
        locations=[[site['lat'], site['lon']], [site['coast_lat'], site['coast_lon']]],
        color='red',
        weight=2.5,
        dash_array='5'
    ).add_to(site_map)

    # Texto da distância
    mid_lat = (site['lat'] + site['coast_lat']) / 2
    mid_lon = (site['lon'] + site['coast_lon']) / 2
    folium.Marker(
        location=[mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 10pt; color : black">{distance:.2f} km</div>',
        )
    ).add_to(site_map)

site_map
```
